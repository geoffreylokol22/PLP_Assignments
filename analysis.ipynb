```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Task 1: Load and Explore the Dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# Task 1: Load and Explore the Dataset
print("First few rows:")
print(df.head())

print("\nData types:")
print(df.dtypes)

print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Task 2: Basic Data Analysis
print("\nBasic statistics:")
print(df.describe())

print("\nMean petal length per species:")
print(df.groupby('species')['petal length (cm)'].mean())
```

```python
# Task 3: Data Visualization
# Line Chart
plt.figure(figsize=(10, 5))
df.groupby('species')['sepal length (cm)'].mean().plot(kind='line', color='darkgreen', linewidth=2)
plt.title('Average Sepal Length by Species', color='darkblue')
plt.ylabel('Sepal Length (cm)', color='darkred')
plt.xlabel('Species', color='darkred')
plt.grid(True, linestyle='--', alpha=0.5)  # Add grid lines
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(x='species', y='petal length (cm)', data=df, hue='species', palette='viridis', width=0.5, legend=False)  # Added hue and legend=False
plt.title('Average Petal Length per Species', color='darkblue')
plt.ylabel('Petal Length (cm)', color='darkred')
plt.xlabel('Species', color='darkred')
plt.tight_layout()
plt.show()
```

```python
# 3. Histogram
plt.figure(figsize=(10, 5))
plt.hist(df['sepal width (cm)'], bins=20, color='skyblue', edgecolor='black')  # Color and edge
plt.title('Distribution of Sepal Width', color='darkblue')
plt.xlabel('Sepal Width (cm)', color='darkred')
plt.ylabel('Frequency', color='darkred')
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines for y-axis
plt.tight_layout()
plt.show()
```

```python
# 4. Scatter Plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df, palette='Set1', s=60)  # Color palette, point size
plt.title('Sepal Length vs. Petal Length', color='darkblue')
plt.xlabel('Sepal Length (cm)', color='darkred')
plt.ylabel('Petal Length (cm)', color='darkred')
plt.tight_layout()
plt.show()
```
